# Retrieval

```python
from cirtorch.networks.imageretrievalnet import init_network
from cirtorch.networks.imageretrievalnet import extract_vectors
from cirtorch.networks.imageretrievalnet import extract_ss, extract_ms
from torchvision import transforms
from cirtorch.utils.general import get_data_root
from torch.utils.model_zoo import load_url
from tqdm import tqdm
import numpy as np
print("Done")
```

```python
import os
os.listdir()
```

```python
import os, glob2
os.chdir('/content/MyDrive/MyDrive/cnnimageretrieval-pytorch/dataset/images')
list_image = glob2.glob('*.jpg')
len(list_image)
```

```python
#Prepare module extraction
PRETRAINED = {
    'rSfM120k-tl-resnet50-gem-w'        : 'http://cmp.felk.cvut.cz/cnnimageretrieval/data/networks/retrieval-SfM-120k/rSfM120k-tl-resnet50-gem-w-97bf910.pth',
    'rSfM120k-tl-resnet101-gem-w'       : 'http://cmp.felk.cvut.cz/cnnimageretrieval/data/networks/retrieval-SfM-120k/rSfM120k-tl-resnet101-gem-w-a155e54.pth',
    'rSfM120k-tl-resnet152-gem-w'       : 'http://cmp.felk.cvut.cz/cnnimageretrieval/data/networks/retrieval-SfM-120k/rSfM120k-tl-resnet152-gem-w-f39cada.pth',
    'gl18-tl-resnet50-gem-w'            : 'http://cmp.felk.cvut.cz/cnnimageretrieval/data/networks/gl18/gl18-tl-resnet50-gem-w-83fdc30.pth',
    'gl18-tl-resnet101-gem-w'           : 'http://cmp.felk.cvut.cz/cnnimageretrieval/data/networks/gl18/gl18-tl-resnet101-gem-w-a4d43db.pth',
    'gl18-tl-resnet152-gem-w'           : 'http://cmp.felk.cvut.cz/cnnimageretrieval/data/networks/gl18/gl18-tl-resnet152-gem-w-21278d5.pth',
}
```

```python
state = load_url(PRETRAINED['gl18-tl-resnet152-gem-w'], model_dir=os.path.join(get_data_root(), 'networks'))
net_params = {}
net_params['architecture'] = state['meta']['architecture']
net_params['pooling'] = state['meta']['pooling']
net_params['local_whitening'] = state['meta'].get('local_whitening', False)
net_params['regional'] = state['meta'].get('regional', False)
net_params['whitening'] = state['meta'].get('whitening', False)
net_params['mean'] = state['meta']['mean']
net_params['std'] = state['meta']['std']
net_params['pretrained'] = False
# network initialization
net = init_network(net_params)
net.load_state_dict(state['state_dict'])
```

```python
# setting up the multi-scale parameters
ms = list(eval('[1, 2**(1/2), 1/2**(1/2)]'))  
print(">>>> Evaluating scales: {}".format(ms))
```

```python
# moving network to gpu and eval mode
net.cuda()
net.eval()
```

```python
# set up the transform
normalize = transforms.Normalize(
    mean=net.meta['mean'],
    std=net.meta['std']
)
transform = transforms.Compose([
    transforms.ToTensor(),
    normalize
])
```

```python
dict_image = {}
with tqdm(total=len(list_image)) as pbar:
  for img in list_image:
    # Feature extraction
    feature = extract_vectors(net, [img], 1024, transform, ms=ms)
    feature = feature.numpy()
    dict_image[img] = feature
    pbar.update(1)
```

```python
with tqdm(total=len(dict_image)) as pbar:
  for img in dict_image:
    # Feature extraction
    np.save(img[:-3] + 'npy', dict_image[img])
    
    pbar.update(1)
```

```python
#@title Khởi tạo Query và Corpus
query = '/content/MyDrive/MyDrive/cnnimageretrieval-pytorch/dataset/images/all_souls_000026.jpg'
corpus = list_image
```

```python
from google.colab.patches import cv2_imshow
import cv2 as cv
query_img = cv.imread(query)
query_img = cv.rectangle(query_img, (78, 5), (714, 1000), (255, 0, 0), 2)
cv2_imshow(query_img)
```

```python
feature_corpus = {}
with tqdm(total=len(corpus)) as pbar:
  for img in corpus:
    feature_corpus[img] = np.load('/content/MyDrive/MyDrive/cnnimageretrieval-pytorch/dataset/images/' +img[:-3] + 'npy',allow_pickle=True)
    pbar.update(1)
len(feature_corpus)
```

```python
feature_query = extract_vectors(net, [query], 1024, transform, bbxs= [(78, 5, 714, 1000)], ms=ms)
vecs = [feature_corpus[i] for i in feature_corpus]
vecs = torch.FloatTensor(vecs)

feature_query = feature_query.numpy()
vecs = vecs.numpy()
```

```python
from numpy.linalg import norm
from numpy import dot
```

```python
#compute cosine distance
cosine_dis = {}
for img in feature_corpus:
  cosine_dis[img] = dot(feature_corpus[img].T, feature_query)/(norm(feature_corpus[img].T)*norm(feature_query))
```

```python
results = sorted(cosine_dis.items(), key = lambda kv:(kv[1], kv[0]))
```

```python
results[-10:]
for i in results[-10:]:
  img = cv.imread(i[0])
  cv2_imshow(img)
```
